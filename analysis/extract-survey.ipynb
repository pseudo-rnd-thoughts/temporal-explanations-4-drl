```python
from typing import NamedTuple


class UserInfo(NamedTuple):
    Survey: int
    UserID: int
    TimeTaken: float
    Gender: str
    Education: str
    DegreeArea: str
    StudiedAI: bool

    BreakoutKnowledge: str
    BreakoutKnowledgeScore: int
    SpaceInvadersKnowledge: str
    SpaceInvadersKnowledgeScore: int
    SeaquestKnowledge: str
    SeaquestKnowledgeScore: int


class IndividualQuestion(NamedTuple):
    Survey: int
    UserID: int
    ExplanationType: str
    QuestionType: str
    QuestionNumber: int
    Environment: str
    Answer: str
    LikertRating: int


class ContrastiveQuestion(NamedTuple):
    Survey: int
    UserID: int
    QuestionType: str
    QuestionNum: str
    Environment: str

    Explanation1: str
    Explanation2: str
    Answer: str
    AnswerNum: int

    OrderedExplanation1: str
    OrderedExplanation2: str
    OrderedAnswer: str
    OrderedAnswerNum: int
```

```python
all_user_info: list[UserInfo] = []

KNOWLEDGE_SCORES = {
    "Never played and have no idea how the game works": 1,
    "Have played once or twice and have minimal knowledge of the game\xa0": 2,
    "Have played and understand the game well": 3,
}

for survey_num, survey in enumerate([survey_1, survey_2]):
    for index, row in survey.iterrows():

        ai_technology_column = survey.columns[9]
        user_info = UserInfo(
            survey_num,
            row["ID"],
            row["Completion time"] - row["Start time"],
            row["Gender"],
            row["Highest of Level of Education"],
            row[
                "If you are completing / have completed a degree, what is the rough area  "
            ],
            row[ai_technology_column] == "Yes",
            row["Breakout"],
            KNOWLEDGE_SCORES[row["Breakout"]],
            row["Space Invaders"],
            KNOWLEDGE_SCORES[row["Space Invaders"]],
            row["Seaquest"],
            KNOWLEDGE_SCORES[row["Seaquest"]],
        )
        all_user_info.append(user_info)

user_info_survey = pd.DataFrame(all_user_info)
user_info_survey.to_csv("anonymised-user-info.csv", index=None)
user_info_survey
```

```python
# Parser individual survey results
individual_survey_results: list[IndividualQuestion] = []

ENVIRONMENTS = [
    "Breakout",
    "SpaceInvaders",
    "Seaquest"
]

INITIAL_COLUMN = 13
FINAL_COLUMN = 78
QUESTIONS_PER_EXPLANATION_TYPE = 13
INDIVIDUAL_LIKERT_RATING = {
    "Strongly Disagree": 1,
    "Disagree": 2,
    "Neutral": 3,
    "Agree": 4,
    "Strongly Agree": 5,
}

SURVEY_1_COLUMNS = list(survey_1.columns)
SURVEY_1_EXPLANATION_ORDER = [
    "Dataset Similarity Explanation",
    "Skill Explanation",
    "Plan Explanation",
    "Perturbation-based Saliency Map",
    "Grad-CAM",
]
for _, row in survey_1.iterrows():
    # Extract individual explanations
    for column_num in range(INITIAL_COLUMN, FINAL_COLUMN):
        if not SURVEY_1_COLUMNS[column_num].startswith("Algorithm "):
            explanation_type = SURVEY_1_EXPLANATION_ORDER[
                (column_num - INITIAL_COLUMN) // QUESTIONS_PER_EXPLANATION_TYPE
            ]
            question_type = re.sub(r"\d+", "", SURVEY_1_COLUMNS[column_num])
            question_index = column_num - INITIAL_COLUMN
            question_num = (
                question_index - 1 - (question_index // QUESTIONS_PER_EXPLANATION_TYPE)
            ) // 4
            # print(index, explanation_type, question, question_num)
            individual_survey_results.append(
                IndividualQuestion(
                    Survey=1,
                    UserID=row["ID"],
                    ExplanationType=explanation_type,
                    QuestionType=question_type,
                    QuestionNumber=question_num,
                    Environment=ENVIRONMENTS[question_num%3],
                    Answer=row[column_num],
                    LikertRating=INDIVIDUAL_LIKERT_RATING[row[column_num]],
                )
            )


SURVEY_2_COLUMNS = list(survey_2.columns)
SURVEY_2_EXPLANATION_ORDER = [
    "Perturbation-based Saliency Map",
    "Grad-CAM",
    "Dataset Similarity Explanation",
    "Skill Explanation",
    "Plan Explanation",
]
for _, row in survey_2.iterrows():
    # Extract individual explanations
    for column_num in range(INITIAL_COLUMN, FINAL_COLUMN):
        if not SURVEY_2_COLUMNS[column_num].startswith("Algorithm "):
            explanation_type = SURVEY_2_EXPLANATION_ORDER[
                (column_num - INITIAL_COLUMN) // QUESTIONS_PER_EXPLANATION_TYPE
            ]
            question_type = re.sub(r"\d+", "", SURVEY_2_COLUMNS[column_num])
            question_index = column_num - INITIAL_COLUMN
            question_num = (
                question_index - 1 - (question_index // QUESTIONS_PER_EXPLANATION_TYPE)
            ) // 4
            # print(index, explanation_type, question, question_num)
            individual_survey_results.append(
                IndividualQuestion(
                    Survey=2,
                    UserID=row["ID"],
                    ExplanationType=explanation_type,
                    QuestionType=question_type,
                    QuestionNumber=question_num,
                    Environment=ENVIRONMENTS[question_num%3],
                    Answer=row[column_num],
                    LikertRating=INDIVIDUAL_LIKERT_RATING[row[column_num]],
                )
            )

individual_survey = pd.DataFrame(individual_survey_results)
individual_survey.to_csv("individual-survey.csv", index=None)
individual_survey
```

```python
# Sanity check
assert len(individual_survey.drop_duplicates(subset=["UserID", "Survey"])) == len(
    survey_1
) + len(survey_2)
assert (
    len(pd.unique(individual_survey["ExplanationType"]))
    == len(SURVEY_1_EXPLANATION_ORDER)
    == len(SURVEY_2_EXPLANATION_ORDER)
)
assert len(pd.unique(individual_survey["QuestionType"])) == 4
assert len(
    individual_survey.drop_duplicates(subset=["Survey", "UserID", "QuestionNumber"])
) * 4 == len(individual_survey)
```

```python
contrastive_survey_results: list[ContrastiveQuestion] = []

CONTRASTIVE_LIKERT_RATING = {
    "This is more true for explanation 1 than explanation 2": 1,
    "This is more true for explanation 2 than explanation 1": 2,
    "This is equally true for explanation 1 and 2": 3,
}
CONTRASTIVE_RATINGS = list(CONTRASTIVE_LIKERT_RATING.keys())
CONTRASTIVE_EXPLANATION = {
    "dataset": "Dataset Similarity Explanation",
    "skill": "Skill Explanation",
    "plan": "Plan Explanation",
    "gradcam": "Grad-CAM",
    "perturbation": "Perturbation-based Saliency Map",
}
CONTRASTIVE_QUESTION_CONVERT = {
    "From the explanation, I understand how the agent works.": "From the explanation, I understand how the agent works.",
    "This explanation of how the agent works has sufficient detail.": "This explanation of how the agent works has sufficient detail.",
    "This explanation of how the agent works is satisfying.": "This explanation of how the agent works is satisfying.",
    "From the explanation, I can predict the agent's next action": "From the explanation, I can predict the agent's next action",
    "\xa0This explanation of how the agent works has sufficient detail.": "This explanation of how the agent works has sufficient detail.",
    "Using the explanation, I can predict the next action": "From the explanation, I can predict the agent's next action",
}

INITIAL_COLUMN = 78
FINAL_COLUMN = len(survey_1.columns)
for survey_num, survey in enumerate([survey_1, survey_2]):

    for _, row in survey.iterrows():
        for column_num, column in enumerate(
            survey_1.columns[INITIAL_COLUMN:FINAL_COLUMN]
        ):
            question_type = CONTRASTIVE_QUESTION_CONVERT[re.sub(r"\d+", "", column)]
            question_num = column_num // 4
            explanation_info = CONTRASTIVE_EXPLANATION_INFO[question_num]
            # print(explanation_info["explanation_1"], explanation_info["explanation_2"], question_num, question_type, row[INITIAL_COLUMN + column_num])

            explanation_1 = CONTRASTIVE_EXPLANATION[explanation_info["explanation_1"]]
            explanation_2 = CONTRASTIVE_EXPLANATION[explanation_info["explanation_2"]]
            answer_num = CONTRASTIVE_LIKERT_RATING[row[INITIAL_COLUMN + column_num]]

            if bool(explanation_1 < explanation_2) != bool(survey_num == 0):  # xor
                ordered_explanation_1, ordered_explanation_2 = (
                    explanation_1,
                    explanation_2,
                )
                ordered_answer_num = answer_num
            else:
                ordered_explanation_1, ordered_explanation_2 = (
                    explanation_2,
                    explanation_1,
                )
                if answer_num == 1:
                    ordered_answer_num = 2
                elif answer_num == 2:
                    ordered_answer_num = 1
                else:
                    ordered_answer_num = 3

            contrastive_survey_results.append(
                ContrastiveQuestion(
                    Survey=survey_num + 1,
                    UserID=row["ID"],
                    QuestionType=question_type,
                    QuestionNum=question_num,
                    Environment=explanation_info["environment"],
                    Explanation1=explanation_1,
                    Explanation2=explanation_2,
                    Answer=row[INITIAL_COLUMN + column_num],
                    AnswerNum=answer_num,
                    OrderedExplanation1=ordered_explanation_1,
                    OrderedExplanation2=ordered_explanation_2,
                    OrderedAnswer=CONTRASTIVE_RATINGS[ordered_answer_num-1],
                    OrderedAnswerNum=ordered_answer_num,
                )
            )

contrastive_survey = pd.DataFrame(contrastive_survey_results)
contrastive_survey.to_csv("contrastive-survey.csv", index=None)
contrastive_survey
```

```python
assert len(contrastive_survey.drop_duplicates(subset=["UserID", "Survey"])) == len(
    survey_1
) + len(survey_2)
assert len(pd.unique(contrastive_survey["QuestionType"])) == 4, pd.unique(
    contrastive_survey["QuestionType"]
)
assert len(pd.unique(contrastive_survey["QuestionNum"])) == 9
assert len(pd.unique(contrastive_survey["Answer"])) == 3
# assert len(contrastive_survey.drop_duplicates(subset=["Survey", "UserID", "QuestionNumber"])) * 4 == len(contrastive_survey)
```
